## Examples of multi axis 3D plot

Geo Dim GC couple

```python
#import IPython
%matplotlib notebook
#%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations

#plt.figure(figsize=(w, h), dpi=d)
#figsize=(15,7.5), dpi= 80

fig = plt.figure(figsize=(15,7.5))
#plt.figure(figsize=(15,7.5), dpi=80)
ax = fig.gca(projection='3d')
ax.set_aspect("equal")

# draw cube
r = [-1, 1]
t = [-1, 2]
u = [-1, 1]
for s, e in combinations(np.array(list(product(r, t, u))), 2):
    if np.sum(np.abs(s-e)) == r[1]-r[0]:
        ax.plot3D(*zip(s, e), color="b")
for s, e in combinations(np.array(list(product(r, t, u))), 2):
    if np.sum(np.abs(s-e)) == t[1]-t[0]:
        ax.plot3D(*zip(s, e), color="b")

# draw sphere
#u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
#x = np.cos(u)*np.sin(v)
#y = np.sin(u)*np.sin(v)
#z = np.cos(v)
#ax.plot_wireframe(x, y, z, color="r")

# draw a point
ax.scatter([-1], [2], [-1], color="g", s=100)

# draw a vector
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d


class Arrow3D(FancyArrowPatch):

    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0, 0), (0, 0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        FancyArrowPatch.draw(self, renderer)

a = Arrow3D([-1.10, 1], [2.10, -1], [-1.1, 1], mutation_scale=20,
            lw=1, arrowstyle="-|>", color="k")
# Demo 2: color
ax.text(-0.3, -1.2, 0, "$R_{GC}/\lambdabar_{e}$", fontsize=24, color='green')
ax.text(-1.5, -2, -1, "$R/\lambdabar_{e}$", fontsize=24, color='black')
ax.text(-1.5, 0.8, 1, "$M^\prime/m_{e}$", fontsize=24, color='green')
ax.text(1.1, 1, -1, "$t/t_{e}$", fontsize=24, color='black')
ax.set_xlabel('Log T')
ax.set_ylabel('Log L')
ax.set_zlabel('Log M')

#ax.text2D(0.05, 0.95, "Geo Dimensional Couple", transform=ax.transAxes)
#ax.text2D(0.05, 0.95, "$M^\prime/m_{e}$", transform=ax.transAxes)
#ax.text2D(0.95, 0.05, "$t/t_{e}$", transform=ax.transAxes)
#ax.text2D(0.02, 0.05, "$R/\lambdabar_{e}$", transform=ax.transAxes)
ax.add_artist(a)
plt.show()
```
